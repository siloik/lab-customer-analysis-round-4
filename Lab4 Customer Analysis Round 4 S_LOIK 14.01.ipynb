```python
file1 = pd.read_csv('marketing_customer_analysis.csv')
```

```python
data = pd.DataFrame(data = file1) 
```

```python
data.head()
```

```python
data.columns = [val.lower().replace(' ','_') for val in data.columns]
print(data.columns)
```

```python
numeric_columns = data.select_dtypes(include = np.number).columns.tolist()
categoric_columns = data.select_dtypes(include = ['object']).columns.tolist()    
print(numeric_columns)
```

```python
df_numeric = pd.DataFrame(data[numeric_columns])
df_object = pd.DataFrame(data[categoric_columns])
df_numeric.head()
```

```python
#check for normality. take every variable and check across 
```

```python
#draw a simple normal distribution
from scipy.stats import norm
import math
normal = norm(0, math.sqrt(9)) 
# print(normal.pdf(4))
# print(normal.cdf(2))       
# print(normal.rvs()) 
fig, ax = plt.subplots(1, 1) #how many rows and columns the plot will have
x = np.linspace(-3,3,1000)
y = norm.pdf(x) 
ax.plot(x,y)
plt.show()
```

```python
#seaborn
for val in df_numeric:
    sns.distplot(df_numeric[val], bins=20) #histplot is now the new one to use
    plt.show()
```

```python
#matplotlib
for val in df_numeric.columns: #iterate through numeric columns 
    plt.hist(df_numeric[val], bins=20)
    plt.show()
```

```python
correlations_matrix = data.corr()
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```
